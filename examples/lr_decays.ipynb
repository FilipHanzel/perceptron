# Learning rate decays
Demo of how learning rate changes with different decay strategies 

```python
import sys
import os

module_path = os.path.abspath("..")
sys.path.append(module_path)

from perceptron.decay import (
    LinearDecay,
    PolynomialDecay,
    TimeBasedDecay,
    ExponentialDecay,
    StepDecay,
)
from matplotlib import pyplot as plt
```

```python
%config InlineBackend.figure_formats = ['svg']
```

```python
base_learning_rate = 1
epochs = 100

linear_decay = LinearDecay(base_learning_rate, epochs)
polynomial_decay = PolynomialDecay(base_learning_rate, epochs, power=2)
time_based_decay = TimeBasedDecay(base_learning_rate, epochs)
exp_decay = ExponentialDecay(base_learning_rate, decay_rate=0.1)
step_decay = StepDecay(base_learning_rate, drop=0.5, interval=10)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot()

ax.plot(range(epochs), [linear_decay(i) for i in range(epochs)], label="linear")
ax.plot(range(epochs), [polynomial_decay(i) for i in range(epochs)], label="polynomial")
ax.plot(range(epochs), [time_based_decay(i) for i in range(epochs)], label="timebased")
ax.plot(range(epochs), [exp_decay(i) for i in range(epochs)], label="exp")
ax.plot(range(epochs), [step_decay(i) for i in range(epochs)], label="step")

ax.grid()
ax.legend()
plt.show()
```
