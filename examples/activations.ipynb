# Activation functions
Demo of how activation functions behave around zero

```python
import sys
import os

module_path = os.path.abspath("..")
sys.path.append(module_path)

from perceptron.activation import (
    Heavyside,
    Linear,
    Relu,
    LeakyRelu,
    Sigmoid,
    Tanh,
)
from matplotlib import pyplot as plt

heavyside = Heavyside()
linear = Linear()
relu = Relu()
leaky_relu = LeakyRelu()
sigmoid = Sigmoid()
tanh = Tanh()

scale = 100
_range = [step / scale for step in range(-5 * scale, 5 * scale)]
```

```python
%config InlineBackend.figure_formats = ['svg']
```

```python
fig = plt.figure(figsize=(20, 7))

ax = fig.add_subplot(2, 3, 1)
ax.plot(_range, heavyside.forward_pass(_range), label="heavyside")
ax.grid(True)
ax.legend(loc="upper left")

ax = fig.add_subplot(2, 3, 2)
ax.plot(_range, linear.forward_pass(_range), label="linear")
ax.grid(True)
ax.legend(loc="upper left")

ax = fig.add_subplot(2, 3, 3)
ax.plot(_range, relu.forward_pass(_range), label="relu")
ax.grid(True)
ax.legend(loc="upper left")

ax = fig.add_subplot(2, 3, 4)
ax.plot(_range, leaky_relu.forward_pass(_range), label="leaky_relu")
ax.grid(True)
ax.legend(loc="upper left")

ax = fig.add_subplot(2, 3, 5)
ax.plot(_range, sigmoid.forward_pass(_range), label="sigmoid")
ax.grid(True)
ax.legend(loc="upper left")

ax = fig.add_subplot(2, 3, 6)
ax.plot(_range, tanh.forward_pass(_range), label="tanh")
ax.grid(True)
ax.legend(loc="upper left")

plt.show()
```
